```python
limit_food_set = []
for i in food_set:
    if i == 'hot_dog' or i=='garlic_bread' or i=='french_fries':
        limit_food_set +=[i]
```

```python
print(limit_food_set)
```

```python
os.mkdir('Train_set')
os.mkdir('Test_set')
```

```python
test_path = os.path.join(curpath,'Test_set')
train_path = os.path.join(curpath,'Train_set')
```

```python
for i in limit_food_set:
    try:
        os.chdir(test_path)
        os.mkdir(i)
        os.chdir(train_path)
        os.mkdir(i)
    except:
        print('Folders already created')
```

```python
import collections
from shutil import copy
def split_data(filepath,src,des):
    classes_images = collections.defaultdict(list)
    
    with open(filepath,'r') as txt:
        paths = [read.strip() for read in txt.readlines()] 
        
        for p in paths:
            food = p.split('/')
        
            if food[0] not in limit_food_set:
                continue
            else:
                classes_images[food[0]].append(food[1] + '.jpg')
        print(classes_images)
    for food in classes_images.keys():
        print('Copying food ',food)
        
        for i in classes_images[food]:
            copy(os.path.join(src,food,i),os.path.join(des,food,i))
    print('Copying Done')
    
```

```python
os.chdir(curpath)
```

```python
print("Creating train data...")
split_data('meta/meta/train.txt',img_path, 'Train_set')
```

```python
print("Creating train data...")
split_data('meta/meta/test.txt',img_path, 'Test_set')
```

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
                rescale = 1./255,
                shear_range = 0.2,
                zoom_range  = 0.2,
                horizontal_flip=True
                )
training_set = train_datagen.flow_from_directory(
                'Train_set',
                target_size=(299,299),
                batch_size = 16,
                class_mode = 'categorical'
                )
```

```python
test_datagen = ImageDataGenerator(
                rescale = 1./255
                )
test_set = train_datagen.flow_from_directory(
                'Test_set',
                target_size=(299,299),
                batch_size = 16,
                class_mode = 'categorical'
                )

```

```python
os.chdir(curpath)
```

```python
food_model = tf.keras.models.Sequential()
```

```python
food_model.add(tf.keras.layers.Conv2D(filters=32,kernel_size=5,activation='relu',input_shape=[299,299,3]))
```

```python
food_model.add(tf.keras.layers.Conv2D(filters=32,kernel_size=(5,5),activation='relu'))
food_model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))
food_model.add(tf.keras.layers.Dropout(rate=0.25))
```

```python
food_model.add(tf.keras.layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
food_model.add(tf.keras.layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
food_model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))
food_model.add(tf.keras.layers.Dropout(rate=0.25))
```

```python
food_model.add(tf.keras.layers.Flatten())
food_model.add(tf.keras.layers.Dense(256, activation='relu'))
food_model.add(tf.keras.layers.Dropout(rate=0.5))
food_model.add(tf.keras.layers.Dense(4, activation='softmax'))
```

```python
food_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
food_model.fit(x=training_set,validation_data=test_set,epochs=25)
```
